```python
import pandas as pd
df=pd.read_csv('/content/netflix_titles.csv')
print(df)
print(df.shape)
df.size
df.tail
df.head

```

**Chart1: Movies VS TV Shows**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Netflix: Movies vs TV Shows Count', fontsize=16)
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

**Chart2:Top 10 Countries**

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with Most Netflix Content', fontsize=16)
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

**Chart3: Release Year Trend**

```python
plt.figure(figsize=(12,6))
df['release_year'].value_counts().sort_index().plot(kind='line', color='red')
plt.title('Netflix Content Release Year Trend', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

**Chart4:Top Genres**

```python
plt.figure(figsize=(10,6))
df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10).plot(kind='barh', color='orange')
plt.title('Top 10 Genres on Netflix', fontsize=16)
plt.xlabel('Count')
plt.gca().invert_yaxis()
plt.show()
```
